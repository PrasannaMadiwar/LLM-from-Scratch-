```python
import tiktoken
```

```python
import importlib
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
text = "hey lvdu ashmmet is gmail id person wjhfiuernjjiwoeh orpnuifo pidok wpfhuerof heuhf ev prasann asn0fen  qi lobve djvb oie ddfr <|endoftext|>"
```

```python
tokens = tokenizer.encode(text,allowed_special={"<|endoftext|>"})
```

```python
print(tokens)
```

```python
string = tokenizer.decode(tokens)
```
