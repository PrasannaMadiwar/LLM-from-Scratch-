```python
import torch
```

```python
from torch.utils.data import Dataset, DataLoader
import tiktoken
```

```python
class GPT_DatasetV1(Dataset):
  def __init__(self,text,tokenizer,max_length,stride):
    self.input = []
    self.output = []

    token_ids = tokenizer.encode(text)

    for i in range(0,len(token_ids)-max_length,stride):
      input_chunk = token_ids[i:i+max_length]
      output_chunk =token_ids[0+i:i+max_length+1]

      self.input.append(torch.tensor(input_chunk))
      self.output.append(torch.tensor(output_chunk))

  def __len__(self):
     return len(self.input)

  def __getitem__(self,idx):
     return self.input[idx], self.output[idx]
```

```python
def create_dataloader(text,batch_size=4,max_length=4,stride=1,shuffle=True,drop_last=True,num_worker=4):
  tokenizer = tiktoken.get_encoding("gpt2")
  dataset = GPT_DatasetV1(text,tokenizer,max_length,stride)

  dataloader = DataLoader(dataset,batch_size=batch_size,shuffle=shuffle,drop_last=drop_last,num_workers=num_worker)
  return dataloader
```

```python
dataloader = create_dataloader(text,batch_size=2,max_length=4,stride=1,shuffle=False,drop_last=True)
```

```python
data_iter = iter(dataloader)
first = next(data_iter)
print(first)
```
