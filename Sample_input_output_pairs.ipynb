```python
with open("/content/simple tokenizer.txt","r",encoding="utf-8") as f:
  text = f.read()
```

```python
import tiktoken
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
tokens = tokenizer.encode(text)
```

```python
max_length = 6
```

```python
x = []
y = []
```

```python
for i in range(1,max_length):
  x.append(tokens[:i])
  y.append(tokens[i])
```

```python
print(x)
print(y)
```

```python
print(tokenizer.decode(x[4]))
print(tokenizer.decode(y))
```
